```python
n=int(input())
l=[int(x) for x in input().split()]
l.sort()
x=int(input())
hm = {}
for ele in l:
    if ele in hm.keys():
        hm[ele] += 1
    else:
        hm[ele] = 1
        

for i in range(len(list(hm.keys()))-1):
    j = i+1
    k = len(hm)-1
    while j<k:
        if list(hm.keys())[i]+list(hm.keys())[j]+list(hm.keys())[k] == x:
            ways = hm[list(hm.keys())[i]]*hm[list(hm.keys())[j]]*hm[list(hm.keys())[k]]
            while(ways):
                print(list(hm.keys())[i], list(hm.keys())[j], list(hm.keys())[k]) 
                ways -= 1            
            j += 1
            k -= 1
        elif list(hm.keys())[i]+list(hm.keys())[j]+list(hm.keys())[k] < x:
            j += 1 
        elif list(hm.keys())[i]+list(hm.keys())[j]+list(hm.keys())[k] > x:  
            k -= 1

                                                                              
                                                                              
                                                                              
                                
        
```

```python

def tsum(l, x):
    hm = {}
    for ele in l:
        if ele in hm.keys():
            hm[ele] += 1
        else:
            hm[ele] = 1
        

    for i in range(len(list(hm.keys()))-1):
        j = i+1
        k = len(hm)-1
        while j<k:
            
            if list(hm.keys())[i]+list(hm.keys())[j]+list(hm.keys())[k] == x:
                
                ways = hm[list(hm.keys())[i]]*hm[list(hm.keys())[j]]*hm[list(hm.keys())[k]]
                while(ways):
                    print(list(hm.keys())[i], list(hm.keys())[j], list(hm.keys())[k]) 
                    ways -= 1            
                j += 1
                k -= 1
            elif list(hm.keys())[i]+list(hm.keys())[j]+list(hm.keys())[k] < x:
                
                j += 1 
            elif list(hm.keys())[i]+list(hm.keys())[j]+list(hm.keys())[k] > x:  
                
                k -= 1

# Main
n=int(input())
l=[int(x) for x in input().split()]
l.sort()
x=int(input())
tsum(l, x)
                                                                              
                                                                              
                                
        
```
