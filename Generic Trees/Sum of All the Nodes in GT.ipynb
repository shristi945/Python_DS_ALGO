```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.children = list()
        

#Take Input from user and make connections 

def TakeInput():
    print("Enter root:")
    rootData = int(input())
    if rootData == -1:
        return 
    root = TreeNode(rootData)
    print("Enter the number of children of", rootData)
    childCount = int(input())
    for i in range(childCount):
        child = TakeInput()
        root.children.append(child)
        
    return root



#print Tree:

def printDetailed(root):
    if root == None:
        return 
    print(root.data, end = ":")
    for child in root.children:
        print(child.data, end = ",")
    print("")
    
    for child in root.children:
        printDetailed(child)
        
#count Nodes

def sumNodes(root):
    if root == None:
        return 0
    sum = root.data
    for child in root.children:
        sum = sum+sumNodes(child)
    return sum


#TakeInput from user
root = TakeInput()
print("Your Tree is:")
printDetailed(root)
print(sumNodes(root))
```
