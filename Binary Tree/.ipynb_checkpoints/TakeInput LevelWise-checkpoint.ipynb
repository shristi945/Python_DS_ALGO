```python
import queue

class BinaryTree:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
    
def printTree(root):
    if root == None:
        return 
    print(root.data,end=":")
    if root.left != None:
        print("L",root.left.data, end = ",")
    if root.right != None:
        print("R",root.right.data, end = "")
    print("")
    printTree(root.left)
    printTree(root.right)
    

def TakeInputLevelWise():
    q = queue.Queue()
    print("Root:")
    rootData = int(input())
    if rootData == -1:
        return 
    root = BinaryTree(rootData)
    q.put(root)
    while not(q.empty()):
        curr = q.get()
        print("Enter Left Child for", curr.data)
        leftChild = int(input())
        if leftChild != -1:
            leftChild = BinaryTree(leftChild)
            curr.left = leftChild
            q.put(curr.left)
            
        print("Enter Right Child for", curr.data)
        rightChild = int(input())
        if rightChild != -1:
            rightChild = BinaryTree(rightChild)
            curr.right = rightChild
            q.put(curr.right)
            
        
        
            
    return root
        
        
#Main

root = TakeInputLevelWise()

#calling Print fun
printTree(root)
```
