```python
class MapNode:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.next = None
class Map:
    def __init__(self):
        self.bucketSize =10
        self.bucketlist = [None for i in range(self.bucketSize)]
        self.count = 0

    def size(self):
        return self.count
        
    def getBucketIndex(self, h):
        return abs(h)%(self.bucketSize)
        
        
    def insert(self, key, value):
        hc = hash(key)
        index = self.getBucketIndex(hc)
        head = self.bucketlist[index]
        while head is not None:
            if head.key == key:
                head.value = value
                return 
            head = head.next
            
        newNode = MapNode(key, value)
        newNode.next = head
        self.bucketlist[index] = newNode
        self.count += 1
        
    def delete(self, key):
        hc = hash(key)
        index = self.getBucketIndex(hc)
        head= self.bucketlist[index]
        prev = None
        while head is not None:
            temp = head
            if head.key == key:
                
                if prev is None:
                    self.bucketlist[index] = head.next
                else:
                    
                    prev.next = head.next
                self.count -= 1
                return temp.key
            head = head.next
        return None
                
    def search(self, key):
        hc = hash(key)
        index = self.getBucketIndex(hc)
        head= self.bucketlist[index]
        
        while head is not None:
            if head.key == key:
                return head.value
            head = head.next
        return None
    

    
    
    
#Main
m = Map()
m.insert(1,5)
print(m.size())
m.insert(2,7)
print(m.size())
m.insert(2,8)
print(m.size())
print(m.delete(1))
print(m.size())
print(m.search(1))

        
        
        
        
```
