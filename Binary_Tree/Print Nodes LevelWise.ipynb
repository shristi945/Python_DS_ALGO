```python
import queue

class BinaryTree:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
    
def printTree(root):
    q = queue.Queue()
    if root == None:
        return 
    q.put(root)
    while (not(q.empty())):
        curr = q.get()
        print(curr.data, end = ":")
        if curr.left is not None:
            print("L:",curr.left.data, end = ",")
            q.put(curr.left)
        else:
            print("L",":",-1, end = ",")
        if curr.right is not None:
            print("R",":",curr.right.data)
            q.put(curr.right)
        print("")
            
    

def TakeInputLevelWise():
    q = queue.Queue()
    print("Root:")
    rootData = int(input())
    if rootData == -1:
        return 
    root = BinaryTree(rootData)
    q.put(root)
    while not(q.empty()):
        curr = q.get()
        print("Enter Left Child for", curr.data)
        leftChild = int(input())
        if leftChild != -1:
            leftChild = BinaryTree(leftChild)
            curr.left = leftChild
            q.put(curr.left)
            
        print("Enter Right Child for", curr.data)
        rightChild = int(input())
        if rightChild != -1:
            rightChild = BinaryTree(rightChild)
            curr.right = rightChild
            q.put(curr.right)
            
        
        
            
    return root
        
        
#Main

root = TakeInputLevelWise()

#calling Print fun
printTree(root)
```
